```python
#파일에서 xtz정보 옆에 있는 아톰의 이름만 골라서 리스트에 담기 (분자량 구하기 위해서)
#파일상 아톰의 정보를 가지고 있는줄은 4번째 줄부터 (4+8)번째 줄까지
```

```python
f= open('one_mol.sdf','r')

mole = f.readlines()
n_atom = mole[3].split()[0]
atom_info = mole[4:4+int(n_atom)]
atom_box = []
for i in atom_info:  #i는 한줄씩 들어옴
    line = i.split()
    print(line[3]) 
    #줄마다 그 인덱싱 출력 인덱싱 값 다르게 넣어보면서 원하는 인덱싱 값 찾기
    atom = line[3]
    atom_box.append(atom)
    
print(atom_box)
```

```python
#분자량 구하기!!
#aw_dic atom의 분자량 딕셔너리
aw_dic = {'H':1,'O':16,'C':12,'F':19,'Cl':35,'N':14,'Br':80}
mw=0
for i in atom_box:
    mw = mw + aw_dic[i]

print(mw)
```
